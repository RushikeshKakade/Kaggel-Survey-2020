## 11. What programming language would you recommend an aspiring data scientist to learn first?

```python
prog_lang  = df['Q8'].value_counts()
prog_lang
```

```python
plt.figure(figsize=(20,7))
plt.xlabel('Programming language for begineer')
plt.ylabel('Count')
plt.title('computing platform use most for data science projects')
prog_lang.plot(kind='bar',color='lime')
```

```python
plt.figure(figsize = (20,15))
plt.xlabel('Programming language for begineer')
prog_lang.plot(kind='pie',autopct='%1.1f%%',textprops={'fontsize': 15})
```
